```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("a3_2021f_coop_tech.csv")
data.head(5)
```

```python
dat = list(data['FALL 2021 Co-op Technologies'])  #  DataFrame to list
dat = pd.Series(dat)                 #  list to Series
det = dat.str.title()                         #   Capitalize texts (Example: Javascript, Visual Studio...)
a = det.value_counts(ascending=True)
```

```python
h = pd.DataFrame(a)                 # All values capitalized 
h2 = pd.DataFrame([len(h[h[0]==1])],index=["Otros"])    # Every frequency of 1 is added
k= pd.concat([h2,h[h[0]>=2]])     # Concatenate all values majores of 1 and the sum of 1 frequencies
```

```python
# Values, Labels and Colors to Plot

labels = list(k.index.values); values = list(k[0]); exp = [0.1]

colores = ["Gold","SkyBlue","Pink","Khaki","LightBlue","PeachPuff","YellowGreen",
           "Azure","LightSalmon"]

for i in range(len(values)-1):(exp.append(0))
```

```python
from IPython.core.pylabtools import figsize

plt.figure(figsize=(800/120, 600/120), dpi=120)
plt.suptitle("Relatively Most Used Technologies",fontsize=14, fontweight= 'black')
plt.pie(values,labels=labels,autopct=lambda pct: ('%i%%' % pct) if pct > 3 else None,
        startangle = 35, explode = exp, rotatelabels=True, shadow= True,
        pctdistance=0.85, colors=colores, textprops = {'fontsize' : 9}) 

plt.text(0, 0.6, 'Many Specialized tecnologies \n Were Reported Only Once', horizontalalignment='center',
     verticalalignment='center', fontsize = 8)

plt.savefig("img.png")
plt.show()
```
